```python
#Jogo da forca
import random
import os

def exibir_forca(erros):
    estagios = [
    # 0 erros
    """
     +---+
         |
         |
         |
        ===
    """,
    # 1 erro
    """
     +---+
     O   |
         |
         |
        ===
    """,
    # 2 erros
    """
     +---+
     O   |
     |   |
         |
        ===
    """,
    # 3 erros
    """
     +---+
     O   |
    /|   |
         |
        ===
    """,
    # 4 erros
    """
     +---+
     O   |
    /|\\  |
         |
        ===
    """,
    # 5 erros
    """
     +---+
     O   |
    /|\\  |
    /    |
        ===
    """,
    # 6 erros (GAME OVER)
    """
     +---+
     O   |
    /|\\  |
    / \\  |
        ===
    """
]
    print(estagios[erros])

def jogar_forca():

    os.system('cls' if os.name == 'nt' else 'clear')

    print(" ðŸ•¸ï¸ðŸ¦‡ Bem-Vindos Ã¡ Forca Assustadora!   ")
    print("Adivinhe o nome de um ser ou objeto de Halloween antes que o **FANTASMA** se complete!\n")

    palavras_halloween = ["ABOBORA", "VAMPIRO", "ZUMBI", "CALDEIRÃƒO", "MORCEGO", "FANTASMA", "CEMITÃ‰RIO"]
    palavra_secreta = random.choice(palavras_halloween)

    letras_certas = []
    letras_erradas = []
    max_erros = 6
    erros = 0

    while erros < max_erros:
        exibir_forca(erros)

        palavra_atual = " "
        for letra in palavra_secreta:
            if letra in letras_certas:
                palavra_atual += letra + " "
            else:
                palavra_atual += "_ "

        print(f"ðŸ‘» Palvra: {palavra_atual}")
        print(f" ðŸ’€ Erros: {erros}/{max_erros}")
        print(f"â˜ ï¸ Letras erradas: {', '.join(letras_erradas)}")

        if "_" not in palavra_atual:
            print(f"\n âœ¨ðŸ§  VITÃ“RIA! A palavra era: **{palavra_secreta}**!")
            return

        tentativa = input("Chute uma letra: ").strip().upper()

        if len (tentativa) != 1 or not tentativa.isalpha():
            print(" ðŸ˜¨ Por favor, chute apenas uma letra!\n")
            continue
        if tentativa in letras_certas or tentativa in letras_erradas:
            print(f" ðŸ‘ï¸ VocÃª jÃ¡ tentou a letra '{tentativa}'. Tente outra.\n")
        elif tentativa in palavra_secreta:
            print(f" ðŸ”¥ Boa! A letra '{tentativa}' estÃ¡ na palavra.")
            letras_certas.append(tentativa)
        else:
            print(f" ðŸ’€ A letra ' {tentativa}' nÃ£o estÃ¡ na palavra. ")
            letras_erradas.append(tentativa) # Adicionado Ã  lista de letras erradas
            erros += 1

        print("-" * 30 + "\n")

    exibir_forca(erros)
    print(f"\n ðŸ”® FIM DE JOGO! O Fantasma te pegou!")
    print(f"A palavra secreta era: **{palavra_secreta}**.")

if __name__=="__main__":
    jogar_forca()
```
